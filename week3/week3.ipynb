```python
from sklearn.svm import SVC
import numpy as np
import pandas
from sklearn import datasets
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

newsgroups = datasets.fetch_20newsgroups(
    data_home='.',
    subset='all',
    categories=['alt.atheism', 'sci.space']
)
```

```python
X = newsgroups.data
y = newsgroups.target

vectorizer = TfidfVectorizer()
X_vect = vectorizer.fit_transform(X)
```

```python
grid = {'C': np.power(10.0, np.arange(-5, 6))}
cv = KFold(n_splits=5, shuffle=True, random_state=241)
clf = SVC(kernel='linear', random_state=241)
gs = GridSearchCV(clf, grid, scoring='accuracy', cv=cv)
gs.fit(X_vect, y)
best_c = gs.best_estimator_.C
print(best_c)
```

```python
# best_c = 1
clf1 = SVC(kernel='linear', random_state=241, C=best_c)
clf1.fit(X_vect, y)
df2 = pandas.DataFrame( np.transpose(clf1.coef_.toarray()), index=np.asarray(vectorizer.get_feature_names()) , columns=["col1"])
df2 = df2.sort_values(by=['col1'])

kys = df2['col1'].keys()
```

```python
top10 = kys[-1:-11:-1]
top10.sort_values()
df2
```

```python
coefs = abs(clf1.coef_.todense().A1)
coefs = np.argsort(coefs)
top10 = coefs[-1:-11:-1]
print(top10)
nms = vectorizer.get_feature_names()
ans = []
for ind in top10:
    ans.append(nms[ind])
print(sorted(ans))
```
