```python
import pandas
data = pandas.read_csv('gbm-data.csv').to_numpy()
```

```python
Xin = data[:, 1::]
Yin = data[:, 0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Xin, Yin, 
                                                    test_size=0.8, 
                                                    random_state=241)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import log_loss
import numpy as np
gbc = GradientBoostingClassifier(n_estimators=250, verbose=True, random_state=241, learning_rate=0.2)
gbc.fit(X_train, y_train)

#     aaa = gbc.staged_decision_function(X_test)
```

```python
logloss_res = []
i = 0
i_min = 0
loss_min = 100
for y_pred in gbc.staged_decision_function(X_test):
    i += 1
    logloss_res.append(log_loss(y_test, 1 / (1 + np.exp(-y_pred))))
    if logloss_res[-1] < loss_min:
        loss_min = logloss_res[-1]
        i_min = i
    
plt.figure()
plt.plot(logloss_res, 'r', linewidth=2)
print(i_min, np.round(loss_min, 2))
fout = open('./ans/3lossmin', 'w', encoding='utf-8')
print(np.round(loss_min, 2), i_min, file=fout, end='')
fout.close()
```

```python
lerates = [1, 0.5, 0.3, 0.2, 0.1]
for lr in lerates:
    gbc = GradientBoostingClassifier(n_estimators=250, verbose=True, random_state=241, learning_rate=lr)
    gbc.fit(X_train, y_train)
    logloss_res = []
    for y_pred in gbc.staged_decision_function(X_test):
        logloss_res.append(log_loss(y_test, 1 / (1 + np.exp(-y_pred))))
    plt.figure()
    plt.plot(logloss_res, 'r', linewidth=2)
```

```python

fout = open('./ans/2qualest', 'w', encoding='utf-8')
print('overfitting', file=fout, end='')
fout.close()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=i_min, random_state=241)
rfc.fit(X_train, y_train)
logloss_res = []
res = log_loss(y_test, rfc.predict_proba(X_test))
# for y_pred in rfc.predict_proba(X_test):
#     logloss_res.append(log_loss(y_test, y_pred))
#     print(y_pred)
fout = open('./ans/4randforest', 'w', encoding='utf-8')
print(np.round(res, 2), file=fout, end='')
fout.close()
```
